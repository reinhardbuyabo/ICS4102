```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# number of time steps
T = 1000

# intial price
P0 = 10

# drift term
mu = 0.001
```

```python
last_p = np.log(P0)

log_returns = np.zeros(T) # array of log returns
prices = np.zeros(T) # array of prices of length T

for t in range(T):
  # sample a random log return (from 0 mean and normal distribution with 0.01 standard deviation)
  r = 0.01 * np.random.randn()

  # compute the new log price = old log price + drift term + random noise
  p = last_p + mu + r

  # store the return and new price
  log_returns[t] = r + mu
  prices[t] = np.exp(p) # exponential, to plot the price and not the log price

  # assign last_p # updating lsast p
  last_p = p
```

```python
plt.figure(figsize=(20,8))
plt.plot(prices)
plt.title("Simulated Prices")
```
